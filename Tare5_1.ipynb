```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import numpy as np

# Cargar el conjunto de datos MNIST
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Capa personalizada para convertir imágenes a escala de grises
class GrayScaleLayer(layers.Layer):
    def call(self, inputs):
        # Asumimos que las imágenes de entrada son en escala de grises (1 canal)
        gray_images = tf.image.grayscale_to_rgb(inputs)
        return gray_images

# Crear una instancia de la capa personalizada
gray_scale_layer = GrayScaleLayer()

# Escoge una imagen de MNIST
image_to_convert = train_images[50]  # Puedes elegir cualquier imagen de entrenamiento

# Añade una dimensión para convertirla en un tensor 4D (lote de imágenes)
image_to_convert = np.expand_dims(image_to_convert, axis=0)
image_to_convert = np.expand_dims(image_to_convert, axis=-1)

# Convierte la imagen a escala de grises
gray_image = gray_scale_layer(image_to_convert)

# Visualiza la imagen original y la imagen a escala de grises
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.imshow(image_to_convert[0, :, :, 0], cmap='gray')
plt.title('Imagen original (escala de grises)')

plt.subplot(1, 2, 2)
plt.imshow(gray_image[0, :, :, 0], cmap='gray')
plt.title('Imagen a escala de grises')

plt.show()
```
